```python

# Implementation of matplotlib function
import numpy as np
import matplotlib.pyplot as plt
  
# First create some toy data:
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)
  
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
  
  
fig.suptitle('matplotlib.pyplot.subplots() Example')
plt.show()
```
