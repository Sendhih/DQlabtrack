```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
boston = load_boston()
boston.keys()
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['MEDV'] = boston['target']
df
```

```python
# np.around(df.describe(), 2)
df.describe().round(2)
```

```python
# df.hist(column='CHAS')
df['CHAS'].plot(kind='hist')
plt.title('CHAS')
plt.show()
```

```python
df['RM'].plot(kind='hist', bins = 20)
plt.title('RM')
plt.show()
```

```python
corr_boston = df.corr().round(2)
corr_boston
```

```python
df.plot('RM','MEDV', kind='scatter', figsize=(8,6))
plt.show()
```

```python
df.plot(x = 'LSTAT', y = 'MEDV', figsize=(8, 6), kind='scatter')
plt.show()
```

BUILD The Model

```python
model = LinearRegression()
```

```python
X = df[['RM']] 
y = df['MEDV']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
# print(X_train.shape)
print(X_test.shape)
# print(y_train.shape)
print(y_test.shape)

# print(X_test)
```

```python
# Fitting Model
model.fit(X_train, y_train)
```

```python
# Find Intercept And Slope (Coeffients)
model.intercept_.round(2)
model.coef_.round(2)
# print(dir(model)) used to see all the attributes
rm = np.array([6.5]).reshape(-1, 1)
rm
```

```python
# Predict all RM per Dwelling
predict= model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, label='Testing data');
plt.plot(X_test.squeeze(), predict, label='Prediction', linewidth=3)
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```
